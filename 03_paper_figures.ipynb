# Figure 2: diagram of more complex simulation flow

```python
fontsize = 0.3 * 72
dx = .2
dy = .2
```

```python
def my_box(i, j, label):
    ax.text(0.1+i*dx, .9-j*dy, '${}$'.format(label),
          ha="center",
          size=fontsize,
          transform=fig1.transFigure,
          bbox=dict(boxstyle='square', fc="w", ec="k", linewidth=2))

def my_dm(i, j, label):
    my_box(i+0, j+0, 'S')
    my_box(i+0, j+1, 'C')

    ax.arrow(i*dx+.09, -j*dy+.9, .005, -0.12, head_width=0.01, head_length=0.02, fc='k', ec='k')
    ax.arrow(i*dx+.11, -j*dy+.89-.12-.04, -.005, 0.12, head_width=0.01, head_length=0.02, fc='k', ec='k')
    ax.arrow(i*dx+.12, -j*dy+.7, .03, -0.005, head_width=0.02, head_length=0.01, fc='k', ec='k')
    
    ax.text(.1 + i*dx, .9-(j+1)*dy-.05, label, ha='center', va='top', fontsize=24)
```

```python
fig1 = plt.figure(figsize=(13, 6))
plt.axis('off')
plt.subplots_adjust(left=0, right=1, bottom=0, top=1)
ax = plt.axes()

width=5
list_of_etiologies= ['ETEC', 'rota', 'shigellosis', 'cholera', 'salmonella', 'EPEC', 'campylobac',
                     'amoeba', 'crypto', 'aeromonas', 'c_diff', 'noro',
                     'adeno']
for z, e in enumerate(list_of_etiologies):
    # HACK: wrap better
    if z >= 9:
        z += 1
        
    i = z % width
    j = z // width
    my_dm(i+j*.5, 1.5*j, e)

plt.savefig('/homes/abie/t/a-{}.png'.format(z))
print(z)
```
